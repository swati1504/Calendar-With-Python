```python
import sys
import calendar
import builtins

n =int(input())
if(type(n) == int and n>-13 and n<13):
    if(n==0):
        c = calendar.TextCalendar(calendar.SUNDAY)
        import datetime
        from datetime import date
        now = datetime.datetime.now()
        str = c.formatmonth(now.year,now.month)
        print(str)

        today = date.today()
        def colored(r, g, b, today):
            return "\033[38;2;{};{};{}m{} \033[38;2;255;255;255m".format(r, g, b, today)

        colored_text = colored(255, 0, 0, today)
        print("Date:",colored_text)
            
    elif (n > 0 and n < 13):
        c = calendar.TextCalendar(calendar.SUNDAY)
        import datetime
        now = datetime.datetime.now()
        str = c.formatmonth(now.year,(now.month+1))
        print(str)
        
        print("Date",end=": ")
        print(now.year,end="-0")
        print(now.month+1,end="-")
        print(now.day)
        
        
    elif(n < 0 and n > -13):
        c = calendar.TextCalendar(calendar.SUNDAY)
        import datetime
        now = datetime.datetime.now()
        str = c.formatmonth(now.year,(now.month-1))
        print(str)
        
        print("Date",end=": ")
        print(now.year,end="-0")
        print(now.month-1,end="-")
        print(now.day)
               
else:
    print("This application should be run with a positive integer, a negative integer or zero. The integer cannot be more than 12 or less than -12.")
    f=open('file1.txt','r')
    f.readline()
    usr = input("Username:")
    pwd = input("Password:")
    for i in f:
        line=i[0:-1].split(',')
        if (line[0] == usr and line[1] == pwd):
            print("Verified")  
        else:
            print("Not Verified")
    f.close()
```
